```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.datasets import load_digits
```

```python
load_digits
```

```python
df=load_digits()
df
```

```python
df.keys()
```

```python
df.target_names
```

```python
df.DESCR
```

```python
df1=pd.DataFrame(df.data)
df1.head()
```

```python
df.images
```

```python
df1['target']=df.target
df1
```

```python
y=df1['target']
y.head()
```

```python
x=df1.iloc[:,:64]
x.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=2)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(xtrain,ytrain)
```

```python
y_pred=model.predict(xtest)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix
cm=np.array(confusion_matrix(y_pred,ytest))
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,ytest)
```

```python
model_C=SVC(C=1)
model_C.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
 model_C=SVC(C=5)
model_C.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
model_kernel=SVC(kernel='linear')
model_kernel.fit(xtrain,ytrain)
model_kernel.score(xtest,ytest)
```

```python
 
model_kernel=SVC(kernel='sigmoid')
model_kernel.fit(xtrain,ytrain)
model_kernel.score(xtest,ytest)
```
